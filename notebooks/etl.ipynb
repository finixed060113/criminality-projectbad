```python
# We import the libraries that we will need for the lab.
import pandas as pd # data handling and dataframes
import numpy as np # array handling and mathematical operations

# allows ignoring Python warnings
import warnings 
warnings.filterwarnings('ignore')
```

```python
try:
    df2010 = pd.read_csv('../data/2010-21crim.csv')
except pd.errors.ParserError:
    print("Se produjo un error al analizar el archivo CSV.")
```

```python
df2010 = df2010.drop(columns=['Unnamed: 13'])
```

```python
try:
    df2022 = pd.read_csv('../data/2022crim.csv')
except pd.errors.ParserError:
    print("Se produjo un error al analizar el archivo CSV.")
```

```python
df2022 = df2022.drop(columns=['Unnamed: 2'])
```

```python
df2022
```

```python

list_year = ["2010","2011","2012","2013","2014","2015","2016","2017","2018","2019","2020","2021"]

name_year = {
    "name": [],
    "2022" : [],
    "2021" : [],
    "2020" : [],
    "2019" : [],
    "2018" : [],
    "2017" : [],
    "2016" : [],
    "2015" : [],
    "2014" : [],
    "2013" : [],
    "2012" : [],
    "2011" : [],
    "2010" : [],
}

for i in range(45, 2385, 45):  # Iterar desde la fila 45 hasta la fila 2384
    fila = df2010["Unnamed: 0"][i]
    name_year["name"].append(fila)
    
for e in  list_year:    
    for i in range(89, 2385, 45):  # Iterar desde la fila 45 hasta la fila 2384
        fila = df2010[e][i]
        name_year[e].append(fila)

for i in range(1,104,2):
    fila = df2022['Enero-diciembre 2022'][i]
    name_year['2022'].append(fila)
```

```python
df = pd.DataFrame(name_year)
df
```
