```python

# We import the libraries that we will need for the lab.
import pandas as pd # data handling and dataframes
pd.options.display.max_columns = None
import numpy as np # array handling and mathematical operations

from geopy.geocoders import Nominatim

# allows ignoring Python warnings
import warnings 
warnings.filterwarnings('ignore')
```

```python
try:
    df2010 = pd.read_csv('../data/datacrims.csv')
except pd.errors.ParserError:
    print("Se produjo un error al analizar el archivo CSV.")
```

```python
df2010 = df2010.drop(columns=['Unnamed: 13'])
```

```python
df2010
```

```python
df = df2010[45:2385]
df = df.reset_index()
df = df.drop(columns=['index'])
```

```python
df
```

```python
list_year = ["2010","2011","2012","2013","2014","2015","2016","2017","2018","2019","2020","2021"]

name_year = {
    "name": [],
    "year": [],
    "infringement": [],
    "values": []
}

for i in range(0, 2340, 45):
    nombre_prov = df["INFRACCIONES"][i]
    for year in list_year:
        for j in range(i + 1, i + 45):
            infringement = df["INFRACCIONES"][j]
            value = df[year][j]
            name_year["name"].append(nombre_prov)
            name_year["year"].append(year)
            name_year["infringement"].append(infringement)
            name_year["values"].append(value)

df1021 = pd.DataFrame(name_year)

df1021
```

```python
# Crear una instancia del geocodificador
geolocator = Nominatim(user_agent="finix")

# Diccionario para almacenar coordenadas en caché
coordenadas_cache = {}

# Función para obtener coordenadas con caché
def sacar_coordenadas_con_cache(nombre):
    if nombre in coordenadas_cache:
        # Si ya tenemos las coordenadas en caché, las retornamos
        return coordenadas_cache[nombre]
    else:
        localizacion = geolocator.geocode(nombre)
        if localizacion:
            # Almacenamos las coordenadas en caché
            coordenadas_cache[nombre] = (localizacion.latitude, localizacion.longitude)
            return coordenadas_cache[nombre]
        else:
            return None, None  # Marcar ubicaciones no encontradas

# Aplicar la función con caché a la columna "name" y crear nuevas columnas para las coordenadas
df1021[['Latitud', 'Longitud']] = df1021['name'].apply(lambda x: pd.Series(sacar_coordenadas_con_cache(x)))

```
